## IMPORT OUR DATA ##

```python

# Naive Bayes On The Iris Dataset vv
from csv import reader

 
# Load a CSV file
def load_csv(filename):
	dataset = list()
	with open(filename, 'r', encoding = 'latin-1') as file:
		csv_reader = reader(file)
		for row in csv_reader:
			if not row:
				continue
			dataset.append(row)
	return dataset
# Import dataset
#dataset = load_csv('spam.csv')
```

```python
import pandas as pd
dataset = pd.read_csv("spam.csv" ,encoding = 'latin-1')
dataset
```

## I-DATA PREPARATION ##

```python
dataset.drop(dataset.columns[[2,3,4]] ,axis = 1, inplace = True)
dataset.head()
```

```python
dataset.columns = ["label", "SMS"]
```

### 1- suppression des ponctuations , espace , casse en minuscule et division des phrases en listes de mots ###

```python
dataset['SMS'] = dataset['SMS'].str.replace('\W+', ' ').str.replace('\s+', ' ').str.strip()
dataset['SMS'] = dataset['SMS'].str.lower()
dataset['SMS'] = dataset['SMS'].str.split()
```

## Vectorisation de l'ensemble de données ##

```python
vocabulary = list(set(dataset['SMS'].sum()))
word_counts_per_sms = pd.DataFrame([
    [row[1].count(word) for word in vocabulary]
    for _, row in dataset.iterrows()], columns=vocabulary)
dataset = pd.concat([dataset.reset_index(), word_counts_per_sms], axis=1).iloc[:,1:]


#word_counts_per_sms
dataset
```

### 2- diviser l’ensemble de données pour former le jeu d'entrainement et le jeu de test ###

```python
train_data = dataset.sample(frac=0.8,random_state=1).reset_index(drop=True)
test_data = dataset.drop(train_data.index).reset_index(drop=True)
train_data = train_data.reset_index(drop=True)
train_data
```

## PROBABILITES A PRIORI DES CLASSES ##

A - HAM

```python
P_ham = Ham_data.shape[0]/train_data.shape[0]
P_ham
```

B - SPAM_probability

```python


P_spam = Spam_data.shape[0]/train_data.shape[0]
P_spam
```

NOMBRE DE MOTS DANS LES MESSAGES SPAMS

```python
Nspam = train_data.loc[train_data['label'] == 'spam','SMS'].apply(len).sum()
Nspam
```

le nombre de mots dans les messages non spam

```python
Nham = train_data.loc[train_data['label'] == 'ham','SMS'].apply(len).sum()
Nham
```

la taille du vocabulaire

```python
Nvoc = len(train_data.columns) - 2
Nvoc
```

## Les résultats ##

```python
alpha = 1
def p_w_spam(word):
    if word in train_data.columns:
        return (train_data.loc[train_data['label'] == 'spam', word].sum() + alpha) / (Nspam + alpha*Nvoc)
    else:
        return 1
      
def p_w_ham(word):
    if word in train_data.columns:
        return (train_data.loc[train_data['label'] == 'ham', word].sum() + alpha) / (Nham + alpha*Nvoc)
    else:
        return 1
```

Fonction de classification 

```python
def classify(message):
    p_spam_given_message = P_spam
    p_ham_given_message = P_ham
    for word in message:
        p_spam_given_message *= p_w_spam(word)
        p_ham_given_message *= p_w_ham(word)
    if p_ham_given_message > p_spam_given_message:
        return 'ham'
    elif p_ham_given_message < p_spam_given_message:
        return 'spam'
    else:
        return 'humain doit classifier'
```

```python
test_data2['predicted'] = test_data2['SMS'].apply(classify)
test_data2
```
